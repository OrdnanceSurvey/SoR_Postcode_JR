```
# Import Libraries
from pyspark.sql import *
from pyspark.sql.types import StringType
from pyspark.sql.functions import lit
```

```
# Load in data
df = table("sorbaseaddressbase.baseblpu")
display(df)
```

```
# Extract relevant information and drop extraneous columns
df = df.withColumn('northing', df['blpu']['northing']).withColumn('easting', df['blpu']['easting']).withColumn('postcode', df['blpu']['postcodeLocator']).withColumn('validFrom', df['blpu']['blpuLifecycle']['startDate']).withColumn('rpc', df['blpu']['rpc'])
columns_to_drop = ['blpu', 'changeType', 'classifications', 'dpas', 'lpis', 'organisations', 'payloadType', 'xRefs', 'PROVENANCE', 'SourceFile']
df_simplified = df.drop(*columns_to_drop)
# Set Valid To to null
df_simplified = df_simplified.withColumn('validTo', lit(None).cast(StringType()))
display(df_simplified)
```

```
# Save (probably don't need this once all in pyspark)
df_simplified.write.saveAsTable("sorbaseaddressbase.postcodes")
```

```
%sql
-- Calculate centroids (probably can do this in pyspark)

CREATE TABLE sorbaseaddressbase.centroids AS
SELECT a.postcode as postcodeUnit, a.validFrom, a.validTo, a.processTimestamp, SUM(a.easting) / COUNT(a.easting) AS centroid_easting, SUM(a.northing) / COUNT(a.northing) AS centroid_northing FROM sorbaseaddressbase.postcodes a
WHERE epoch == 66 AND rpc == 1 OR rpc == 2
GROUP BY a.postcode, a.processTimestamp, a.validFrom, a.validTo;
```

```
%python
# Create custom WKT conversion

def convertToWKT(easting, northing):
  result_string = 'POINT({0} {1})'.format(easting, northing)
  return result_string

udfToWKT = udf(convertToWKT, StringType())
```

```
# Add WKT Geometry Column
df_centroids = table('sorbaseaddressbase.centroids')
df_geom = df_centroids.withColumn('geometry', udfToWKT(df_centroids['centroid_easting'], df_centroids['centroid_northing'])).drop('centroid_easting').drop('centroid_northing')
print('{0} postcode centroids found.'.format(df.count()))
df_geom.write.saveAsTable("sorbaseaddressbase.output")
```
